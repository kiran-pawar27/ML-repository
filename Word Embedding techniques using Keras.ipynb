```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
sent = ['the glass of milk',
        'the glass of juice',
        'I am a good boy',
        'I am a good Developer',
        'understand the meaning of words',
        'your videos are good']
```

```python
sent
```

```python
# one hot representation of words

vocab_size = 10000

one_hot_rep = [one_hot(words,vocab_size) for words in sent]
print(one_hot_rep)
# it gives index from dictionary
```

### one hot encoding words pass through embedding layer of keras. it produces words embedding

```python
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
```

```python
# padding makes each sentence with same lenght by adding zeros pre or post
sent_length =8
embedd_docs = pad_sequences(one_hot_rep,padding='pre',maxlen=sent_length)
```

```python
print(embedd_docs)
```

```python
dim=15
```

```python
model =Sequential()
model.add(Embedding(vocab_size,10,input_length = sent_length))
model.compile('adam','mse')
```

```python
model.summary()
```

```python
print(model.predict(embedd_docs))
```
